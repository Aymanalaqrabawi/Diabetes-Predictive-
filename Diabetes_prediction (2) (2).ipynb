```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
diabetes_df = pd.read_csv(r"Healthcare-Diabetes.csv")
diabetes_df.head()
```

```python
diabetes_df.shape
```

```python
diabetes_df.info()
```

```python
diabetes_df=diabetes_df.drop_duplicates()
```

```python
diabetes_df.columns
```

```python
diabetes_df['Outcome'].value_counts()
```

```python
diabetes_df = diabetes_df.drop(columns=["Id"])
```

```python
correlation = diabetes_df.corr()
sns.heatmap(correlation, annot=True, cmap="RdBu")
```

```python
Y = diabetes_df["Outcome"]
X = diabetes_df.drop("Outcome",axis=1)
```

```python
X.head()
```

```python
Y.head()
```

```python
from sklearn.model_selection import train_test_split,cross_val_score
x_train, x_test, y_train, y_test = train_test_split(X,Y,test_size=0.2, random_state=0)
```

```python
x_train.shape, y_train.shape
```

```python
x_test.shape, y_test.shape
```

```python
from sklearn.ensemble import BaggingClassifier
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.linear_model import LogisticRegression
logistic_parameters = {'C': 1, 'penalty': 'l2'}

BC_model = LogisticRegression(**logistic_parameters, max_iter=5000)
BC_model.fit(x_train, y_train)

```

```python
diabetes_df.columns
```

```python
y_pred = BC_model.predict(x_test)
from sklearn.metrics import accuracy_score, precision_score, confusion_matrix
print("Accuracy : ",accuracy_score(y_test,y_pred))
print("Precision : ",precision_score(y_test,y_pred))
print("confusion matrix : \n",confusion_matrix(y_test,y_pred))
```

```python

import matplotlib.pyplot as plt

sns.kdeplot(y_test, label="actual Values", fill=True)
sns.kdeplot(y_pred, label="Predicted Values", fill=True)

plt.ylim(0,)  
plt.legend()
plt.title("Distribution of True vs Predicted Values")
plt.xlabel("Value")
plt.ylabel("Density")
plt.show()
```

```python
import pickle

with open('new2_Diabetes_model.pkl', 'wb') as f:
    pickle.dump(BC_model,f)
```
